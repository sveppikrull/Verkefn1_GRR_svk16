```python
n = 1000000
niter = 1000
```

Þau föll sem notuð eru til að finna vendisstökin auk falls sem býr til drefingu B.

```python
def Random_pivot(n):
    return random.randint(0,n)

def Median_3(arr):
    n = arr.shape[0]
    mid = math.ceil(n/2)
    if(arr[0] <= arr[mid] <= arr[n-1]):
        return mid
    if(arr[n-1] <= arr[mid] <= arr[0]):
        return mid
    if(arr[0] <= arr[n-1] <= arr[mid]):
        return n-1
    if(arr[mid] <= arr[mid] <= arr[0]):
        return n-1
    if(arr[mid] <= arr[0] <= arr[n-1]):
        return 0
    if(arr[n-1] <= arr[0] <= arr[mid]):
        return 0

def Mom_pivot(arr,k):
    m = math.ceil(arr.shape[0]/k)
    vals = np.zeros(m)
    for i in range(m):
        vals[i] = statistics.median(np.array(arr[i*k:(i+1)*k]))
    if(np.random.rand()>0):
      mom_val = statistics.median_low(vals)
    else:
      mom_val = statistics.median_high(vals)
    mom_idx = np.where(arr == mom_val)[0][0]
        
    return mom_idx

def light_shuffle(arr):
  n = arr.shape[0]
  for i in range(n):
    if(100<=i<=(n-100)-10):
      swap_idx = np.random.randint(i-100,high = (i + 100))
    elif(i<100):
      swap_idx = np.random.randint(0,high =100)
    else:
      swap_idx = np.random.randint(i - 100,high =(n-1))   
    tmp  = arr[i]
    arr[i] = arr[swap_idx]
    arr[swap_idx] = tmp 
  return arr
```

#Gagnasöfnun
Fundið er staðsetningu vendisstaksins í röðuðu fylki fyrir allar þrjár aðferðirnar og þeim niðurstöðum komið fyrir í fylki.

```python
MOM_A = []
Random_A = []
Three_A = []

arr = np.arange(n)
for i in range(niter):
  random.shuffle(arr)
  mom_piv = Mom_pivot(arr,5)
  three_piv = Median_3(arr)
  random_piv = Random_pivot(arr.shape[0])

  MOM_A.append(rankdata(arr, method='ordinal')[mom_piv])
  Three_A.append(rankdata(arr,method='ordinal')[three_piv])
  Random_A.append(rankdata(arr,method='ordinal')[random_piv])
  if(i%100==0):
    print("iteration: ", i)

print("\nMOM\nIterations: ",len(MOM_A)," Mean pivot placement: ",np.mean(MOM_A))
print("\nThree Median\nIterations: ",len(Three_A)," Mean pivot placement: ",np.mean(Three_A))
print("\nRandom Pivot\nIterations: ",len(Random_A)," Mean pivot placement: ",np.mean(Random_A))


MOM_A = np.array(MOM_A)
Random_A = np.array(Random_A)
Three_A = np.array(Three_A)

savetxt('MOM_A.csv', MOM_A, delimiter=',')
savetxt('Random_A.csv', Random_A, delimiter=',')
savetxt('Three_A.csv', Three_A, delimiter=',')
```

```python
MOM_B = []
Random_B = []
Three_B = []

for i in range(int(niter/2)): #Var að renna út á tíma og hver keyrsla var tímafrek
  arr = np.arange(n)
  arr = light_shuffle(arr)

  mom_piv = Mom_pivot(arr,5)
  three_piv = Median_3(arr)
  random_piv = Random_pivot(arr.shape[0])

  MOM_B.append(rankdata(arr, method='ordinal')[mom_piv])
  Three_B.append(rankdata(arr,method='ordinal')[three_piv])
  Random_B.append(rankdata(arr,method='ordinal')[random_piv])
  if(i%100==0):
    print("iteration: ", i)

print("\nMOM\nIterations: ",len(MOM_B)," Mean pivot placement: ",np.mean(MOM_B))
print("\nThree Median\nIterations: ",len(Three_B)," Mean pivot placement: ",np.mean(Three_B))
print("\nRandom Pivot\nIterations: ",len(Random_B)," Mean pivot placement: ",np.mean(Random_B))


MOM_B = np.array(MOM_B)
Random_B = np.array(Random_B)
Three_B = np.array(Three_B)

savetxt('MOM_B.csv', MOM_B, delimiter=',')
savetxt('Random_B.csv', Random_B, delimiter=',')
savetxt('Three_B.csv', Three_B, delimiter=',')
```
